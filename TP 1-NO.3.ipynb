```python
#2502133660_Adam Zadit Taqwa
import matplotlib.pyplot as plt

# Data negara dan jumlah penduduk
countries = ['Indonesia', 'Brunei Darussalam', 'Malaysia', 'Thailand', 'Vietnam']
population = [267, 5.63, 327, 126, 69.4]

# Membuat figure dan axis untuk line chart
plt.figure(figsize=(10, 6))  # Ukuran grafik (width, height)

# Plotting line chart 
plt.plot(countries, population, marker='o', linestyle='-', color='b', linewidth=2, markersize=8, markerfacecolor='red')

# Menambahkan judul dan label
plt.title('Jumlah Penduduk di Beberapa Negara ASEAN', fontsize=16, fontweight='bold')
plt.xlabel('Negara', fontsize=14)
plt.ylabel('Jumlah Penduduk (Juta)', fontsize=14)

# Menambahkan grid 
plt.grid(True, which='both', axis='both', linestyle='--', linewidth=0.7, alpha=0.7)

# Menambahkan anotasi untuk setiap titik data
for i, txt in enumerate(population):
    plt.annotate(f'{txt}', (countries[i], population[i]), textcoords="offset points", xytext=(0,5), ha='center', fontsize=10)

# Mengubah tampilan sumbu X dan Y
plt.xticks(fontsize=12, rotation=45)  # Memutar label X agar lebih terbaca
plt.yticks(fontsize=12)

# Menampilkan chart
plt.tight_layout()  # Mengatur layout agar tidak ada yang terpotong
plt.show()
```
